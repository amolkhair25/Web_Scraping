```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.1mg.com/drugs-all-medicines?page=1&label=a"
headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.5938.150 Safari/537.36"}
page=requests.get(url, headers=headers)
soup=BeautifulSoup(page.content,'lxml')
soup
```

```python
meds=soup.find_all(class_="style__flex-1___A_qoj")
len(meds)
```

```python
first=meds[0]
first
```

```python
fn=first.find(class_="style__font-bold___1k9Dl style__font-14px___YZZrf style__flex-row___2AKyf style__space-between___2mbvn style__padding-bottom-5px___2NrDR").get_text()
fn
```

```python
fn1, pr=fn.split("MRP₹")
fn1, pr
```

```python
pres=first.find(class_="style__rx___3pKXG style__font-12px___2ru_e").get_text()
pres
```

```python
info=first.find(class_="style__flex-column___1zNVy style__font-12px___2ru_e")
info1=[f.get_text() for f in info.find_all(class_="style__padding-bottom-5px___2NrDR")]
info1
```

```python
cont=first.find(class_="style__font-12px___2ru_e style__product-content___5PFBW style__display-inline-block___2y7gd").get_text()
cont
```

```python
name, price, prescp, infos, conts = [], [], [], [], []
for m in meds:
    fn1, pr1=m.find(class_="style__font-bold___1k9Dl style__font-14px___YZZrf style__flex-row___2AKyf style__space-between___2mbvn style__padding-bottom-5px___2NrDR").get_text().split("MRP₹")
    name.append(fn1)
    price.append(pr1)
    prescp.append(m.find(class_="style__rx___3pKXG style__font-12px___2ru_e").get_text())
    infos.append([f.get_text() for f in m.find(class_="style__flex-column___1zNVy style__font-12px___2ru_e").find_all(class_="style__padding-bottom-5px___2NrDR")])
    conts.append(m.find(class_="style__font-12px___2ru_e style__product-content___5PFBW style__display-inline-block___2y7gd").get_text())
len(name), len(price), len(prescp), len(infos), len(conts)
```

```python
d={"Med Name":[i for i in name[::2]], "Price":[i for i in price[::2]], "Prescription":[i for i in prescp[::2]], "Information":[i for i in infos[::2]], "Contents":[i for i in conts[::2]]}
df=pd.DataFrame(d)
df
```

```python
name, price, prescp, infos, conts = [], [], [], [], []
alp="abcdefghijklmnopqrstuvwxyz"
for ch in alp:
    for j in range(1,35):
        url=f"https://www.1mg.com/drugs-all-medicines?page={j}&label={ch}"
        page=requests.get(url, headers=headers)
        soup=BeautifulSoup(page.content,'lxml')

        meds=soup.find_all(class_="style__flex-1___A_qoj")

        for m in meds:
            fn1, pr1=m.find(class_="style__font-bold___1k9Dl style__font-14px___YZZrf style__flex-row___2AKyf style__space-between___2mbvn style__padding-bottom-5px___2NrDR").get_text().split("MRP₹")
            name.append(fn1)
            price.append(pr1)
            prescp.append(m.find(class_="style__rx___3pKXG style__font-12px___2ru_e").get_text())
            infos.append([f.get_text() for f in m.find(class_="style__flex-column___1zNVy style__font-12px___2ru_e").find_all(class_="style__padding-bottom-5px___2NrDR")])
            conts.append(m.find(class_="style__font-12px___2ru_e style__product-content___5PFBW style__display-inline-block___2y7gd").get_text())

```

```python
len(name), len(price), len(prescp), len(infos), len(conts)
```

```python
d1={"Med Name":[i for i in name[::2]], "Price":[i for i in price[::2]], "Prescription":[i for i in prescp[::2]], "Information":[i for i in infos[::2]], "Contents":[i for i in conts[::2]]}
df1=pd.DataFrame(d1)
df1
```
